```python
from cogent3 import get_app

loader = get_app("load_aligned", format="fasta", moltype="dna")
aln = loader("data/primate_brca1.fasta")
model = get_app("model",
    "CNFGTR",
    tree="data/primate_brca1.tree",
    optimise_motif_probs=True,
)
result = model(aln)
result
```
